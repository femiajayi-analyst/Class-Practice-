```python
#reading our dataset
import pandas as pd
covid_data = pd.read_csv('COVID clinical trials.csv')
covid_data
```

```python
#removing the columns that won't be useful for the analysis
import pandas as pd
covid_data.drop(columns = ['NCT Number', 'Title', 'Acronym', 'Outcome Measures', 'Sponsor/Collaborators', 'Study Designs', 'Other IDs',
                      'Primary Completion Date','First Posted','Results First Posted','Last Update Posted','Study Documents',
                      'URL'])
```

```python
#Checking for any duplicate values across the dataset
covid_data.duplicated().any()
```

```python
#Checking for any missing values across the dataset
covid_data.isnull().sum()
```

```python
#to check which start date has the highest counts so we can take our missing values from it
covid_data.groupby('Start Date')['Start Date'].count()
```

```python
#extracting values from April 1, 2020 to replace the 34 missing values across other start date
covid_data.groupby('Start Date')['Start Date'].count()
covid_data[['Start Date']] = covid_data[['Start Date']].fillna(value='April 1, 2020')
```

```python
#To crosscheck if our missing values has been replaced
covid_data.isnull().sum()
```

```python
#To check which start date has the highest counts after replacing the missing values
covid_data['Start Date'].value_counts()
```

```python
#To count how many enrollment will have in each start date
covid_data.groupby('Start Date')['Enrollment'].value_counts()
```

```python
#to know the number of study type we have in each start date
covid_data.groupby('Start Date')['Study Type'].value_counts()
```
